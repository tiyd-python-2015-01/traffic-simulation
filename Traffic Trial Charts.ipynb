```python
from trafficsim.car import Car
from trafficsim.simulation import Simulation
import seaborn
import matplotlib.pyplot as plt
import statistics as st
import numpy as np
```

```python
%matplotlib inline
```

```python
normal_sim = Simulation()
```

```python
speeds, positions = normal_sim.simulate()
```

```python
times = [[x for x in range(30)] for x in range(240)] 
```

# Normal Mode Trial

```python
plt.figure(figsize=(9, 9))
plt.scatter(positions, times, color='black', marker=',', alpha=.12)
plt.ylabel('Time')
plt.xlabel('Position')
plt.show()
```

# Hard Mode Trial

```python
hard_sim = Simulation(210, 7000)
hard_speeds, hard_positions = hard_sim.simulate()
hard_times = [[x for x in range(210)] for x in range(240)] 
```

```python
plt.figure(figsize=(9, 9))
plt.scatter(hard_positions, hard_times, color='black', marker=',', alpha=.12)
plt.ylabel('Time')
plt.xlabel('Position')
plt.show()
```

# Nightmare Mode Trial

```python
nightmare_sim = Simulation(210, 7000, True)
nightmare_speeds, nightmare_positions = hard_sim.simulate()
nightmare_times = [[x for x in range(210)] for x in range(240)] 
```
