# Traffic Simulation

Assumptions

Drivers want to go up to 120 km/hr.
The average car is 5 meters long.
Drivers want at least 4 car lengths between them and the next car.
Drivers will accelerate 2 m/s up to their desired speed as long as they have room to do so.
If another car is too close, drivers will match that car's speed until they have room again.
If a driver would hit another car by continuing, they stop.

Drivers will randomly (10% each second) slow by 2 m/s.
This section of road will have a new car enter it every 2-6 seconds.
This section of road is one lane going one way.

Given all this information, create a simulation of traffic on this road. 

Even though the road is not circular in real life, you should treat it as such: cars exiting the road start again at the beginning. This is to simulate a continuous stream of traffic. 

When you start the simulation, add 30 cars to the road per kilometer, evenly spaced. Then run the simulation for one minute to get a continuous, randomized stream of traffic.

```python
def rotate(a_list):
    if len(a_list) == 0:
        return a_list
    return a_list[1:] + a_list[:1]
```

```python
class Simulation:
    """
    Responsibilities:
    * Places the cars on the road
    * Adjust cars positions & speeds based on car proximity
    * iterating over turns
    * reporting data
    
    Collaborators:
    - Car
    """
    
    def __init__(self, road_length = 1000):
        self.car_list = []
        self.road_length = road_length
        self.distance_list = []#this was more for testing purposes at first, prob not necessary
        self.iterations = 0
        
    def generate_cars(self, num_cars):
        """Generates cars, places them evenly apart, and creates a car pairs list
        of tuples"""
        position = 0
        for _ in range(num_cars):
            #position += round((self.road_length / num_cars), 2)#this works but didn't seem to be rounding properly
            position += 33
            self.car_list.append(Car(position))
        self.car_list = list(zip(self.car_list, rotate(self.car_list)))##thank you for your help with this one Clinton
        return self.car_list
    
    def random_slow(self):
        if random.random() < .1:
            return True
        else:
            return False    
    
#     def make_pairs(self):
#         for cars in self.car_list:
#             list_pair = list(zip(self.car_list, rotate(self.car_list)))
#         return list_pair

#     def get_distance_test(self):        
#             """ checks distance between the car pairs and appends value to distance_list"""
#             for pair in self.car_list:
#                 distance = (pair[1].position - pair[0].position)
#                 if distance >= 0:
#                     self.distance_list.append(distance)
#                 else:
#                     overlap = self.road_length % abs(pair[1].position - pair[0].position)
#                     self.distance_list.append(overlap)
#             return self.distance_list    


    def get_distance(self, pair):
        """checks distance between the car pairs and returns the absolute value in meters"""
        distance = (pair[1].position - pair[0].position)
        if distance >= 0:
            return distance
        else:
            #this is in here if one car is behind start/finish line and the other one is over it...Example 980m & 20m
            overlap = self.road_length % abs(pair[1].position - pair[0].position)
            return overlap
        
 
    def speed_adjust(self):
        """Updates the cars new speed for next move. 10% chance of slow down instead"""
        for car in self.car_list:
            #slows down if too close to next car...4 car lengths + length of car ahead of it
            if self.get_distance(car) <= (car[1].length * 5):
                car[0].speed = car[1].speed
            elif car[0].speed > 2 and self.random_slow():
                car[0].speed -= 2
            elif car[0].speed >= car[0].max_speed:
                car[0].speed == car[0].max_speed #Not sure if this is correct...does it need = or ==?
            else:
                car[0].speed += 2
        return car[0].speed
    
    
    def move_cars(self):
        """Moves cars based on proximity and speed of the one in front of it."""
        for car in self.car_list:
            #this is for if a car would rear-end the car ahead of it on the next move
            if ((car[0].position + car[0].speed) % self.road_length) >= ((car[1].position + car[1].speed) % self.road_length):
                car[0].position = ((car[1].position + car[1].speed) - 25) % self.road_length
                car[0].speed = 0###DO I STOP THE CAR OR MATCH SPEED TO CAR IN FRONT?
            else:
                car[0].position = ((car[0].position + car[0].speed) % self.road_length)
        return car[0].position
    
        
    def iterate(self):
        self.speed_adjust()
        self.move_cars()
        self.iterations += 1

    def report(self):
        return self.iterations, self.car_list

###I couldnt get this to move the cars so I used what you did with the roomba above    
#     def traffic_run(self):
#         position_list = []
#         self.speed_adjust()
#         self.move_cars()
#         position_list.append(self.car_list)
#         return position_list
#         #return self.car_list

    
```
