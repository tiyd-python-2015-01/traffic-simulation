```python
import matplotlib.pyplot as plt
import seaborn as sb
import statistics 

%matplotlib inline
```

```python
from simulation import Simulation
from road import Road
from car import Car
import statistics as st
import numpy as np
```

```python
def run_simulation(cars,desired_speed,starting_speed,trials):
    average_speed = []
    for _ in range(trials):
        road = Road()
        simulation = Simulation(road)
        simulation.make_cars(cars,desired_speed,starting_speed)
        simulation.start()
        for car in simulation.traffic:
            average_speed.append(car.speed)

    return st.mean(average_speed)
```

```python
avg_33 = []
for _ in range(1000):
    avg_33.append(run_simulation(30,33,0,1))
```

```python

plt.hist(avg_33, bins=30)
plt.xlabel("Meters per second")
plt.ylabel("Frequency")
plt.title("Average travelling speed 33m/s")

ymin, ymax = plt.ylim()
mean = statistics.mean(avg_33)
stdev = statistics.pstdev(avg_33)
plt.vlines(mean, ymin, ymax, color="red")
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, linestyles="dashed", color="red")
plt.show()
```
