```python
from car import Car
from road import Road
from simulation import Simulation
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python
%matplotlib inline
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
plt.rcParams['savefig.dpi'] = 2 * plt.rcParams['savefig.dpi']
```

```python
sim = Simulation()
```

```python
x = 1000  # iterations
car_speeds_array = np.zeros(shape=(x,30))
for i in range(x):
    sim.run_simulation(120)
    car_speeds = np.array(sim.get_car_speeds())
    car_speeds_array[i] = car_speeds
```

```python
car_position_array, car_speed_array = sim.run_simulation(120)  # car positions for each second for one trial
```

```python
mean_car_speed_per_trial = car_speeds_array.mean(axis=1)
```

# Mean final speed for each trial

```python
plt.scatter(range(1000),mean_car_speed_per_trial)
plt.xlim(0,1000)
plt.xlabel("trial")
plt.ylabel("speed")
plt.show()
```

# Standard deviation of each trial

```python
plt.scatter(range(1000), car_speeds_array.std(axis=1))
plt.xlim(0,1000)
plt.xlabel("trial")
plt.ylabel("standard deviation of each trial")
plt.show()
```
